>Using the data below, answer the following questions:

>1) Which entities (top 5) had the largest population density in 2020?
>2) Which entities have more water area than land area?
>3) Which entities increased in population the most in the last 10 years?
>4) What state bird accounts for the largest population as of 2020? Land area?
>5) How many entities' largest city is their capital city?
>6) Which city has the largest percent drop from their largest city to their 5th largest? 100*(1st largest - 5th largest)/(1st largest)

```python
import pandas as pd
```

```python
facts = pd.read_csv('../data/state_facts.tsv',delimiter="\t")
facts.head(5)
```

```python
#1 Which entities (top 5) had the largest population density in 2020?
facts['Pop_Density']=facts['Pop_2020']/facts['Area_land']
facts1 = facts.sort_values(by='Pop_Density',ascending=False)
facts1.head(5)
```

```python
#2 Which entities have more water area than land area?
facts2=facts[facts['Area_water']>facts['Area_land']]
facts2
```

```python
#3 Which entities increased in population the most in the last 10 years?
facts['Pop_inc']=facts['Pop_2020']-facts['Pop_2010']
facts3=facts.sort_values(by='Pop_inc',ascending=False)
facts3.head(5)
```

```python
#4 What state bird accounts for the largest population as of 2020? Land area?
facts4=facts.sort_values(by='Pop_2020',ascending=False)
print(facts4['State_bird'].head(1))

#California quail
#Willow ptarmigan

```

```python
facts5=facts.sort_values(by='Area_land',ascending=False)
print(facts5['State_bird'].head(1))
```

```python
#5 How many entities' largest city is their capital city?
count = 0
facts['Capital=LargestCity'] = facts['Capital'] == facts['City_1']
for i in range(56):
    if facts['Capital=LargestCity'][i] == True:
        count += 1
count
                                     
```

```python
##6 Which city has the largest percent drop from their largest city to their 5th largest? 
facts['LargestPercentDrop'] = ((facts['city_1_pop'] - facts['city_5_pop'])*100)/facts['city_1_pop']
facts6 = facts.sort_values(by='LargestPercentDrop', ascending=False)
facts6.head(1)
```

```python
date = pd.read_csv('../data/state_dates.tsv',delimiter="\t")
final = pd.merge(facts, date, left_on='USPS_code',right_on='Abbreviation',how='outer')
final
```

```python
#7 Of the states that joined the United States before 1790, what is the most common state flower?
final['Date_Fixed'] = pd.to_datetime(final['Date'], errors='raise')
facts7 = final[final['Date_Fixed'] < pd.datetime(1790,1,1)]
facts7['State_flower'].value_counts().head(1)
```
